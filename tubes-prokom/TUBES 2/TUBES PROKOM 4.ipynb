# ## Visualisasi data
# perbandingan kategori

```python
fig_size = plt.rcParams['figure.figsize']
fig_size[1] = 6
fig_size
```

```python
# total loan berdasarkan region
# data dikelompokkan berdasarkan region untuk jumlah pinjaman total, lalu dijumlahkan semua data untuk masing-masing region
# selanjutnya, hasil pengelompokan dan penjumlahan langsung diplotkan
graph1 = loan2.groupby('Region')['Total Loan Amount in Mio US$'].sum()
graph1.plot(kind='bar',title='total loan berdasarkan region')
```

```python
# statistical loan berdasarkan sector
# data dikelompokkan berdasarkan sector untuk jumlah pinjaman total, lalu dijumlahkan semua data untuk masing-masing sector
# selanjutnya, hasil pengelompokan dan penjumlahan langsung diplotkan
graph1 = loan2.groupby('Sector')['Statistical Amount in Mio US$'].sum()
graph1.plot(kind='barh',title='statistical loan berdasarkan sector')
```

```python
# data dikelompokkan berdasarkan region untuk dua buah data, lalu dijumlahkan semua data untuk masing-masing region
# selanjutnya, hasil pengelompokan dan penjumlahan langsung diplotkan
graph = loan.groupby('Region').sum()
graph.plot(kind='bar', y=['Total Loan Amount in Mio US$', 'Statistical Amount in Mio US$'],title='Perbandingan total loan dan statstical loan dari berbagai region')
```

# # Analisis Regresi

```python
# melihat grafik total loan
loan.loc[loan['Region'] == 'Sub-Saharan Africa'].plot(kind='scatter', x='Total Loan Amount in Mio US$', y='Statistical Amount in Mio US$')
```

```python
# regresi linear dengan library sklearn
from sklearn.linear_model import LinearRegression

x = np.array(loan2.loc[:,'Total Loan Amount in Mio US$']).reshape(-1,1)
y = np.array(loan2.loc[:,'Statistical Amount in Mio US$']).reshape(-1,1)

model = LinearRegression()
model.fit(x, y)
x_predict = 1000
y = model.predict(x_predict)
y
```

```python
loan2['Total Loan Amount']
```

```python
# plotting regresi linear

xfit = np.linspace(loan2['Total Loan Amount in Mio US$'].min(), loan2['Total Loan Amount in Mio US$'].max()).reshape(-1,1)
yfit = model.predict(xfit)

plt.scatter(x, y)
plt.plot(xfit, yfit, color='green')
plt.show()
```
