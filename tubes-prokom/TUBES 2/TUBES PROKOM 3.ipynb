# Analisis Regresi

```python
# regresi linear dengan library sklearn
from sklearn.linear_model import LinearRegression

indonesia = oil2.loc[oil2['LOCATION'] == 'IDN']

x = np.array(indonesia.loc[:,'TIME']).reshape(-1,1)
y = np.array(indonesia.loc[:,'Value']).reshape(-1,1)

model = LinearRegression(fit_intercept = True)
model.fit(x, y)
x_predict = 2016
y_predict = model.predict(x_predict)
y_predict
```

```python
# plotting regresi linear

xfit = np.linspace(oil2['TIME'].min(), oil2['TIME'].max()).reshape(-1,1)
yfit = model.predict(xfit)

plt.scatter(x,y)
plt.plot(xfit, yfit, color='green')
```
