## Visualisasi data

### Perbandingan katogori

```python
# total unit terjual berdasarkan region
graph1 = sales_rec.groupby('Region')['Units Sold'].sum()
graph1.plot(kind='bar')
```

```python
# rata-rata unit terjual berdasarkan negaranya
graph1 = sales_rec.groupby('Item Type')['Units Sold'].mean()
graph1.plot(kind='barh')
```

```python
graph = sales_rec.groupby('Region').sum()
graph.plot(kind='bar', y=['Total Revenue', 'Total Profit'])
```

### Perubahan terhadap waktu

```python
# unit sold seiring tanggal shipping
# catatan: data diurutkan dulu berdasarkan ship date berurut naik, kemudian diplot

sales_rec.sort_values('Ship Date').plot(kind='line', x='Ship Date', y='Total Profit')
```

```python
# menampilkan grafik penjualan region Asia seiring waktu

sales_rec.sort_values('Ship Date').loc[sales_rec['Region'] == 'Asia'].plot(kind='line', x='Ship Date', y='Units Sold')
```

```python
# menampilkan grafik penjualan sayur-mayur seiring waktu shipping

sales_rec.sort_values('Ship Date').loc[sales_rec['Item Type'] == 'Vegetables'].plot(kind='line', x='Ship Date', y='Units Sold')
```

```python
# menampilkan grafik semua barang secara terpisah seiring waktu

sales_rec.sort_values('Ship Date').groupby('Item Type')['Ship Date', 'Total Profit'].plot(kind='line', x='Ship Date')
```

# Analisis Regresi

```python
# melihat tabel penjualan sereal
sales_rec.loc[sales_rec['Item Type'] == 'Cereal']
```

```python
# melihat grafik penjualan sereal
sales_rec.loc[sales_rec['Item Type'] == 'Cereal'].plot(kind='scatter', x='Units Sold', y='Total Profit')
```

```python
# regresi linear dengan library sklearn
from sklearn.linear_model import LinearRegression

cereal = sales_rec.loc[sales_rec['Item Type'] == 'Cereal']

x = np.array(cereal.loc[:,'Units Sold']).reshape(-1,1)
y = np.array(cereal.loc[:,'Total Profit']).reshape(-1,1)

model = LinearRegression(fit_intercept = True)
model.fit(x, y)
x_predict = 1000
y_predict = model.predict(x_predict)
y_predict
```

```python
# plotting regresi linear

xfit = np.linspace(0, sales_rec['Units Sold'].max()).reshape(-1,1)
yfit = model.predict(xfit)

plt.scatter(x,y)
plt.plot(xfit, yfit, color='green')
```
